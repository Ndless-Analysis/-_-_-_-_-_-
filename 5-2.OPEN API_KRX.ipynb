## 1. open api 불러오기

```python
# import 모듈: 모듈 전체 불러오기
# from 모듈 import 이름: 모듈내에서 필요한 것만 가져오기 

import urllib.request
import urllib.parse       # = from urllib import urlencode, quote_plus, unquote
```

```python
# URL인코딩=퍼센트 인코딩
# quote(): 한글-->문자 URL 인코딩
# unquote(): 문자(%)-->한글(단일문자) URL 디코딩

decode_key= unquote('J8lmKiLNXc9XWmTXjAUV5k9y9r2rQp%2BX3sW%2BrDCcShcWyUsFJEhg1T2uHV6ZuTVwneT0K0E%2BLdQFf0dCGd9xMw%3D%3D')
print(decode_key)
```

```python
# 서비스 URL = 기본 URL(오퍼레이션명까지)

url= 'http://apis.data.go.kr/1160100/service/GetSecuCompInfoService/getSecuCompGeneInfo'
```

```python
# URL 파라미터 분리
# 샘플데이터 그대로 적어도 되고 바꿔도 됨
# 여기서 numOfRows를 전체 데이터 550개로 함

queryParams = '?' + urlencode({quote_plus('ServiceKey'): decode_key,
                               quote_plus('pageNo'): '1',
                               quote_plus('numOfRows'): '550',
                               quote_plus('resultType'): 'xml',
                               quote_plus('title'): '증권_일반현황_임직원현황',
                               quote_plus('basYm'): '200512'})
```

```python
# requests 모듈: HTTP(GET, POST메소드) 요청을 보내는 모듈 / API를 사용할 때 주로 이용
# GET사용 시 requests.get()사용

import requests
response= requests.get(url+queryParams).text
```

```python
import bs4
from lxml import html
xmlobj= bs4.BeautifulSoup(response, 'lxml-xml')
```

```python
print(xmlobj)
```

## 2. 데이터 불러오기

### 2-1. 데이터 분리

```python
# xml데이터는 body-items-item 태그로 구분되어 있음
# item 태그로 분리

rows= xmlobj.findAll('item')
print(rows)
```

```python
# 0번 행 조회

rows[0]
```

```python
# 0번 행(rows[0])의 모든 값을 하나의 column값으로 보고 리스트에 담기

columns= rows[0].find_all()
columns
```

```python
# 리스트 조회하기

columns[0]
columns[0].name
columns[0].text
```

### 2-2. 반복문으로 만들기

```python
rowList= []
nameList= []
columnList= []

columnsLen= len(columns)
for j in range(0,columnsLen):
    eachColumn= columns[j].text
    columnList.append(eachColumn)
rowList.append(columnList)
```

```python
# 전체 데이터 값에 대한 반복문

import pandas as pd

rowList = []
nameList = []
columnList = []

rowsLen = len(rows)
for i in range(0, rowsLen):
    columns = rows[i].find_all()
    columnsLen = len(columns)
    for j in range(0, columnsLen):

        if i == 0:
            nameList.append(columns[j].name)
   
        eachColumn = columns[j].text
        columnList.append(eachColumn)
    rowList.append(columnList)
    columnList = []    
    
result = pd.DataFrame(rowList, columns=nameList)
```
