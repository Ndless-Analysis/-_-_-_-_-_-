```python
import urllib.request as ul
import xmltodict
import json
import sys
import io
import pandas as pd
from collections import OrderedDict
import os
import threading
import csv
from datetime import datetime
import time
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
nanum = pd.read_csv('C:\\★팀프로젝트\\데이터수집\\데이터\\나눔ID.csv', encoding = "EUC-KR")
```

```python
# 쏘카
# 0~400 추출
# 400~450 추출
# 457번째 인덱스 오류
# 458~ 추출
# 예외 처리를 하여 오류나는 행은 무시

allcnt = []
cnt = []
nm = []
date_list =[]
time_list =[]
    
for i in nanum.거점ID[0:len(nanum)]:
    ur = "http://openapi.seoul.go.kr:8088/4c68514c6267616e36345749576b69/xml/NanumcarCarList/1/5/"
    so = "/so"
    
    try:
        url = ur + str(i) + so
        request = ul.Request(url)
        response = ul.urlopen(request)
        responseData = response.read()
        rD = xmltodict.parse(responseData)
        
        if rD["NanumcarCarList"]["RESULT"]['CODE'] == "INFO-000":
                allcnt.append(rD["NanumcarCarList"]["row"]["reservAbleAllCnt"])
                cnt.append(rD["NanumcarCarList"]["row"]["reservAbleCnt"])
                nm.append(rD["NanumcarCarList"]["row"]["SPONAM"])
                
                date_time = str(now.year) + '-' + str(now.month) + '-' + str(now.day)
                time = str(now.hour) + ":"+ str(now.minute)+":"+str(now.second)
                
                date_list.append(date_time)
                time_list.append(time)
                
    except Exception as e1:
        continue

        
allcnt2 = pd.DataFrame(allcnt,columns=['allcnt'])
cnt2 = pd.DataFrame(cnt,columns=['cnt'])
nm2 = pd.DataFrame(nm,columns=['name'])
now = datetime.now() 

#time데이터프레임 만들기
#time = str(now.month) + str(now.day) + "_" + str(now.hour) + ":"+ str(now.minute)+":"+str(now.second)
#time_list.append(time)

date_df =  pd.DataFrame(date_list,columns=['datetime'])
time_df= pd.DataFrame(time_list,columns=['Time'])

rr = pd.concat([allcnt2,cnt2,nm2,date_df,time_df],1)

#파일저장명
strNow = str(now.month) + str(now.day) + "_" + str(now.hour) + "_"+ str(now.minute)
#strNow
len(rr)
na = "C:\\★★팀프로젝트_수정\\데이터수집\\나눔카현황\\socar\\so_"
no = strNow +"_"+str(len(rr))+ ".csv"
na + no
rr.to_csv(na+no, mode='a', encoding="EUC-KR")
```
