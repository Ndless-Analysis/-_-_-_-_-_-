```python
# 데이터 불러오기

import pandas as pd
df= pd.read_csv('C:\\data\\fin_data.csv')
df.head()
```

```python
# 컬럼명 바꾸기

df.columns=['기준년월','법인등록번호','금융회사코드','금융회사명','임직원수','임직원구분코드','임직원구분코드명']
```

```python
df
```

```python
# 금융회사별 임직원 수 더하기

fin_cnt_sum= df.groupby(by=['금융회사명'], as_index=False).sum()
fin_cnt_sum
```

```python
# 임직원수 별로 정렬하기

asc= fin_cnt_sum.sort_values(by=['임직원수'], ascending=False)
asc
```

```python
# 증권사 제외한 상위 10개 추출

asc_max10= asc[1:11]
asc_max10
```

```python
# 임직원수가 가장 많은 최상위 10개 회사 시각화 (막대 그래프)

import matplotlib.pyplot as plt
import seaborn as sns

plt.bar(asc_max10.금융회사명, asc_max10.임직원수, align='edge', color='orange', edgecolor='gray')
plt.title('상위 10개 회사 임직원 수 현황', fontsize=20)
plt.xlabel('금융회사명', fontsize=13)
plt.ylabel('임직원수', fontsize=13)
plt.xticks(rotation = 45 )
plt.show()
```
