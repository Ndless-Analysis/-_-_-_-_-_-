# 1. 연체현황 - 연체건수, 연체 차주 수

```python
업권명 = df['업권명'].unique()
```

```python
for i in range(0,len(업권명)):
    #fig = plt.figure()

    ax = df[df.업권명 == 업권명[i]].plot(x='연월', y='차주수', kind='line')
    plt.title(업권명[i],fontsize = 15)
    plt.show()

    fig = ax.get_figure()
    fig = ax.get_figure()
    ax.figure.savefig('C:/Users/ksy/Desktop/개인/코로나19 팀프로젝트/1.현황분석/png/연체현황/차주수/'+str(i)+'.'+업권명[i]+'_연도별_차주수.png', bbox_inches='tight')
```
